```python
import matplotlib.pyplot as plt
import numpy as np
import heapq

INF = float('inf')

def shortest_path(graph, start, end):
    n = len(graph)
    distance = [INF] * n
    previous = [-1] * n
    distance[start] = 0

    heap = [(0, start)]

    while heap:
        dist, node = heapq.heappop(heap)

        if dist > distance[node]:
            continue

        for neighbor, weight in graph[node].items():
            new_dist = dist + weight

            if new_dist < distance[neighbor]:
                distance[neighbor] = new_dist
                previous[neighbor] = node
                heapq.heappush(heap, (new_dist, neighbor))

    path = []
    current = end

    while current != -1:
        path.append(current)
        current = previous[current]

    return path[::-1]



nodes = [(1,1), (1,5), (4,5), (4,8), (8,8), (12,5), (12,1)]
names = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
x_coordinates = [x*10 for x, _ in nodes]
y_coordinates = [y*10 for _, y in nodes]
plt.scatter(x_coordinates, y_coordinates, marker='s', s=40)

for i, name in enumerate(names):
    plt.text(x_coordinates[i], y_coordinates[i], name, ha='center', va='bottom')


path_pair = [[(10, 10),(10, 50)], [(10, 50),(40, 50)], [(40, 50),(40, 80)], [(40, 80),(80, 80)], [(80, 80),(120, 50)], [(120, 50),(120, 10)], [(120, 10),(10, 10)], [(40, 50),(120, 50)], ]

for path in path_pair:
    plt.plot([path[0][0], path[1][0]], [path[0][1], path[1][1]], color="orange")

sx = [20,60,100,100]
sy = [30,70,70,20]

stationsMap = {}

# for i in range(len(sx)):
#     for path in path_pair:
#
#



plt.scatter(sx,sy,color="green")
for i in range(4):
    plt.text(sx[i],sy[i], str(f"station{1+i}"), ha= "left")



graph = {
    0: {1: 40, 6: 110},
    1: {0: 40, 2: 30},
    2: {1: 30, 3: 30, 5: 80},
    3: {2: 30, 4: 40},
    4: {3: 40, 5: 50},
    5: {2: 80, 4: 50, 6: 40},
    6: {0: 110, 5: 40}
}




startingNode = 0
endNode = 4

passing_nodes = shortest_path(graph, startingNode, endNode)

for i in range(1, len(passing_nodes)):
    l = passing_nodes[i-1]
    c = passing_nodes[i]
    x1 = np.linspace(10*nodes[l][0], 10*nodes[c][0], graph[l][c]//3)
    y1 = np.linspace(10*nodes[l][1], 10*nodes[c][1], graph[l][c]//3)
    plt.scatter(x1, y1, color="red", s=50)
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Plot')
plt.show()
```
